```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
fig = plt.figure()
x = np.linspace(0, 10, 1000) #generates an array of 1000 points from 0 to 10
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
#colors can be chosen by name, short code, hex code, or grayscale between 0 and 1
#closer to 0 is darker, closer to 1 is lighter
plt.plot(x, np.sin(x), color='green')
plt.plot(x, np.cos(x), color='b')
plt.plot(x, np.sin(x+1), color='#ff0000')
plt.plot(x, np.cos(x+1), color='0.75')
```

```python
#line styles: solid, dashed, dashdot, dotted
plt.plot(x, np.sin(x), color='green', linestyle='solid')
plt.plot(x, np.cos(x), color='b', linestyle='dashed')
plt.plot(x, np.sin(x+1), color='#ff0000', linestyle='dashdot')
plt.plot(x, np.cos(x+1), color='0.75', linestyle='dotted')
```

```python
#shorter version of linestyles
plt.plot(x, np.sin(x), color='green', linestyle='-') #solid
plt.plot(x, np.cos(x), color='b', linestyle='--') #dashed
plt.plot(x, np.sin(x+1), color='#ff0000', linestyle='-.') #dashdot
plt.plot(x, np.cos(x+1), color='0.75', linestyle=':') #dotted
```

```python
#combine color and linestyle to make it even shorter
#must use single character color code?
plt.plot(x, np.sin(x), '-g') #solid green
plt.plot(x, np.cos(x), '--b') #dashed blue
plt.plot(x, np.sin(x+1), '-.r') #dashdot red
plt.plot(x, np.cos(x+1), ':c') #dotted cyan
```
