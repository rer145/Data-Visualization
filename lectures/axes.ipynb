```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')
```

```python
fig = plt.figure()
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), '-b')
plt.plot(x, np.cos(x), '--r')
```

```python
plt.plot(x, np.sin(x), '-b')
plt.plot(x, np.cos(x), '--r')
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
#reversing axes is possible
plt.plot(x, np.sin(x), '-b')
plt.plot(x, np.cos(x), '--r')
plt.xlim(11, -1)
plt.ylim(1.5, -1.5)
```

```python
plt.plot(x, np.sin(x), '-b')
plt.plot(x, np.cos(x), '--r')
plt.axis([-0.2, 10.2, -1.1, 1.1]) #set all limits in one command
```

```python
#create plot figure
plt.figure()

#creates a panel with axis of 2, 1, 1 (rows, columns, panel #)
plt.subplot(2, 1, 1) 
plt.plot(x, np.sin(x))

#create second panel
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
#creating an array of plots
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
#generate a plot grid
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
plt.xlim(0, 10.0)
plt.ylim(0, 10.0)
ax = plt.axes()
```

```python
#adding a legend
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```
