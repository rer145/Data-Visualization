```python
#inline contains static images in notebook
#could also use %matplotlib notebook for use inside notebook (interactive)

%matplotlib inline  
import matplotlib.pyplot as plt
import numpy as np
```

```python
#list available styles
plt.style.available
```

```python
plt.style.use('seaborn-colorblind')
```

```python
#create array of 100 numbers between 0 and 10
x = np.linspace(0, 10, 100)
type(x)
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
#can save as png, jpeg, pdf, svg, etc.
fig.savefig('first_plot.png')

#list out available options
fig.canvas.get_supported_filetypes()
```

```python
#show image in notebook
from IPython.core.display import Image
Image('first_plot.png')
```

```python
#another way to display
#opens in new window/application as different file
from PIL import Image
my_fig = Image.open('first_plot.png').show()
```

```python
#another way to display
import matplotlib.image as mpimg
img = mpimg.imread('first_plot.png')
plt.imshow(img)
plt.show()
```
