```python
import pandas as pd
```

```python
jt2016 = pd.read_csv('rround-2016.txt', sep=';', low_memory=False, encoding='latin-1')
jt2017 = pd.read_csv('rround-2017.txt', sep=';', low_memory=False, encoding='latin-1')
```

```python
field2017 = jt2017.loc[(jt2017['Player Name'] != 'Thomas, Justin')]
jt2017 = jt2017.loc[(jt2017['Player Name'] == 'Thomas, Justin')]
jt2016 = jt2016.loc[(jt2016['Player Name'] == 'Thomas, Justin')]
```

```python
jt = pd.concat([jt2016, jt2017])
jt = jt[(jt["Round Score"] >= 58)]
jt.describe()
```

```python
list(jt)
```

```python
jt.fillna(0)
field2017.fillna(0)
```

```python
def print_stat_compare(field):
    print('JT 2017 ' + field + ':', jt.loc[(jt['Tournament Year'] == 2017)][field].mean())
    print('JT 2016 ' + field + ':', jt.loc[(jt['Tournament Year'] == 2016)][field].mean())
    print('Field 2017 ' + field + ':', field2017[field].mean())
    print('\n')
```

# Off The Tee

```python
jt['Driving Accuracy'] = jt['Driving Acc. %(Fairways Hit)'] / jt['Driving Acc. %(Possible Fairways)']
jt['Driving Average'] = jt['Driving Dist. - All Drives(Tot. Dist.)'] / jt['Driving Dist. - All Drives(# of Drives)']
field2017['Driving Accuracy'] = field2017['Driving Acc. %(Fairways Hit)'] / field2017['Driving Acc. %(Possible Fairways)']
field2017['Driving Average'] = field2017['Driving Dist. - All Drives(Tot. Dist.)'] / field2017['Driving Dist. - All Drives(# of Drives)']
```

```python
print_stat_compare('Driving Accuracy')
print_stat_compare('Driving Average')
```

# Approach Shots

```python
jt['Approach 50-75 Average'] = jt['App. 50-75 yds(ft)'] / jt['App. 50-75 yds(attempts)']
jt['Approach 75-100 Average'] = jt['App. 75-100 yds(ft)'] / jt['App. 75-100 yds(attempts)']
jt['Approach 100-125 Average'] = jt['App. 100-125 yds(ft)'] / jt['App. 100-125 yds(attempts)']
jt['Approach 50-125 Average'] = jt['Approaches 50-125 Yards(ft)'] / jt['Approaches 50-125 Yards(attempts)']
jt['Approach 125-150 Average'] = jt['Approaches 125-150 Yards(ft)'] / jt['Approaches 125-150 Yards(attempts)']
jt['Approach 150-175 Average'] = jt['Approaches 150-175 Yards(ft)'] / jt['Approaches 150-175 Yards(attempts)']
jt['Approach 175-200 Average'] = jt['Approaches 175-200 Yards(ft)'] / jt['Approaches   175-200 Yards(attempts)']
jt['Approach >200 Average'] = jt['Approaches >200 Yards(ft)'] / jt['Approaches >200 Yards(attempts)']

field2017['Approach 50-75 Average'] = field2017['App. 50-75 yds(ft)'] / field2017['App. 50-75 yds(attempts)']
field2017['Approach 75-100 Average'] = field2017['App. 75-100 yds(ft)'] / field2017['App. 75-100 yds(attempts)']
field2017['Approach 100-125 Average'] = field2017['App. 100-125 yds(ft)'] / field2017['App. 100-125 yds(attempts)']
field2017['Approach 50-125 Average'] = field2017['Approaches 50-125 Yards(ft)'] / field2017['Approaches 50-125 Yards(attempts)']
field2017['Approach 125-150 Average'] = field2017['Approaches 125-150 Yards(ft)'] / field2017['Approaches 125-150 Yards(attempts)']
field2017['Approach 150-175 Average'] = field2017['Approaches 150-175 Yards(ft)'] / field2017['Approaches 150-175 Yards(attempts)']
field2017['Approach 175-200 Average'] = field2017['Approaches 175-200 Yards(ft)'] / field2017['Approaches   175-200 Yards(attempts)']
field2017['Approach >200 Average'] = field2017['Approaches >200 Yards(ft)'] / field2017['Approaches >200 Yards(attempts)']
```

```python
print_stat_compare('Approach 50-75 Average')
print_stat_compare('Approach 75-100 Average')
print_stat_compare('Approach 100-125 Average')
print_stat_compare('Approach 50-125 Average')
print_stat_compare('Approach 125-150 Average')
print_stat_compare('Approach 150-175 Average')
print_stat_compare('Approach 175-200 Average')
print_stat_compare('Approach >200 Average')
```

```python
jt['GIR Average'] = jt['Total Greens in Regulation'] / jt['Total Holes Played']
field2017['GIR Average'] = field2017['Total Greens in Regulation'] / field2017['Total Holes Played']
```

# Going for the Green

```python
jtGFG2017Total = jt.loc[(jt['Tournament Year'] == 2017)]['Going for Green(attempts)'].sum() + jt.loc[(jt['Tournament Year'] == 2017)]['Going for Green(non-attempts)'].sum()
jtGFG2017Attempts = jt.loc[(jt['Tournament Year'] == 2017)]['Going for Green(attempts)'].sum()
jtGFG2017Success = jt.loc[(jt['Tournament Year'] == 2017)]['Going for the Green(successes)'].sum()
jtGFG2017Average = jtGFG2017Success / jtGFG2017Attempts

jtGFG2016Total = jt.loc[(jt['Tournament Year'] == 2016)]['Going for Green(attempts)'].sum() + jt.loc[(jt['Tournament Year'] == 2016)]['Going for Green(non-attempts)'].sum()
jtGFG2016Attempts = jt.loc[(jt['Tournament Year'] == 2016)]['Going for Green(attempts)'].sum()
jtGFG2016Success = jt.loc[(jt['Tournament Year'] == 2016)]['Going for the Green(successes)'].sum()
jtGFG2016Average = jtGFG2017Success / jtGFG2016Attempts

fieldGFG2017Total = field2017['Going for Green(attempts)'].sum() + field2017['Going for Green(non-attempts)'].sum()
fieldGFG2017Attempts = field2017['Going for Green(attempts)'].sum()
fieldGFG2017Success = field2017['Going for the Green(successes)'].sum()
fieldGFG2017Average = fieldGFG2017Success / fieldGFG2017Attempts
```

```python
print('JT 2017 GFG Average: {0} ({1} of {2} chances)'.format(jtGFG2017Average, jtGFG2017Attempts, jtGFG2017Total))
print('JT 2016 GFG Average: {0} ({1} of {2} chances)'.format(jtGFG2016Average, jtGFG2016Attempts, jtGFG2016Total))
print('Field 2017 GFG Average: {0} ({1} of {2} chances)'.format(fieldGFG2017Average, fieldGFG2017Attempts, fieldGFG2017Total))
```
