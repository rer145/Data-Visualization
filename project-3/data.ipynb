```python
jt2016 = data2016.loc[(data2016['Player Name'] == 'Thomas, Justin')]
jt2017 = data2017.loc[(data2017['Player Name'] == 'Thomas, Justin')]
```

```python
jt2017 = jt2017[(jt2017["Round Score"] >= 58)]
jt2017.loc[(jt2017["Round Score"] <= 58)]
```

```python
jt2017["Round Score"].describe()
```

```python
jt2017majors = jt2017[(jt2017["Event Name"] == "Masters Tournament") |
                    (jt2017["Event Name"] == "U.S. Open") |
                     (jt2017["Event Name"] == "The Open Championship") |
                     (jt2017["Event Name"] == "PGA Championship")]
jt2017majors.describe()
```

```python
jt2016majors = jt2016[(jt2016["Event Name"] == "Masters Tournament") |
                    (jt2016["Event Name"] == "U.S. Open") |
                     (jt2016["Event Name"] == "The Open Championship") |
                     (jt2016["Event Name"] == "PGA Championship")]
jt2016majors.describe()
```

```python
# compare tee time start with round score
# short game analysis
# off the tee analysis
# putting analysis
```

```python
jt2017['TeeTimeHour'] = 0
jt2016['TeeTimeHour'] = 0
jt2017['TeeTimeHour'] = jt2017['Tee Time'].str[0:2]
jt2016['TeeTimeHour'] = jt2016['Tee Time'].str[0:2]
```

```python
jt2017['TeeTimeHour'] = jt2017['TeeTimeHour'].astype(int)
jt2016['TeeTimeHour'] = jt2016['TeeTimeHour'].astype(int)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
jt2017.groupby(['TeeTimeHour'])['TeeTimeHour'].count()
```

```python
agg = jt2017.groupby(['TeeTimeHour'])['TeeTimeHour'].count()
agg.plot.bar()
```

```python
agg = jt2017.groupby(['TeeTimeHour'])['Round Score'].mean()
agg.plot.bar()
```

```python
agg = jt2016.groupby(['TeeTimeHour'])['Round Score'].mean()
agg.plot.bar()
```

```python
fig = plt.figure()
plt.plot(jt2016.groupby(['TeeTimeHour'])['Round Score'].mean(), label='2016')
plt.plot(jt2017.groupby(['TeeTimeHour'])['Round Score'].mean(), label='2017')
plt.legend()
plt.show()
```

```python
jt = pd.concat([jt2016, jt2017])
jt.head()
```
